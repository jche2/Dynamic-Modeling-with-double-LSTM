```python
import bagpy
from bagpy import bagreader
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
import numpy as np
from torch import nn
import os
```

## Load All bag files

```python
bag_file_path = "/home/jingche/Desktop/argo/data/lstm_vehicle/"
files = os.listdir(bag_file_path)
bag_files = [bag_file_path + f for f in files if f.endswith(('.bag'))]
bag_files.sort()
```

## Transfer all the bag file and some of the topics to CSV

```python
ctrl_msg_csv_dir = []
loc_msg_csv_dir = []
for i in range(len(bag_files)):
    bag = bagreader(bag_files[i])
    ctrl_msg_csv_dir.append(bag.message_by_topic("/vehctrl_msg"))
    loc_msg_csv_dir.append(bag.message_by_topic("/location_msg"))
```
